```python
import pandas as pd
import re
```

```python
data=[]

with open("ai-ticket/pyre/function.counts") as fi :
  for x in fi:
    aa = re.split(r'[\s,]+', x)
    if aa[1]:
      func = aa[2]
      parts= func.split(".")
      dd = dict(count=int(aa[1]),
              function =  func,
              dots=len(parts),
              path1=parts[0],pathn=parts[-1])
      data.append(dd)
```

```python
df= pd.DataFrame(data)
```

```python
df
```

```python
table = pd.pivot_table(df,
        values='count',
        index=['path1'], sort=True,
        # columns=['function'],
      aggfunc="sum")
```

```python
table.sort_values(by="count")
```
